```python
import torchio as tio
import nibabel as nib
import numpy as np
from pathlib import Path
from PIL import Image
```

```python
segmentation_dir = Path('stent_segmentations')
labels = dict()
unique_labels = set()
for segmentation_path in segmentation_dir.glob('*.nii'):
    img = nib.load(segmentation_path)
    labels[segmentation_path.name] = tio.LabelMap(tensor=img.get_fdata()[np.newaxis, ...])
    a = np.unique(img.get_fdata()).tolist()
    unique_labels = unique_labels.union(set(a))
    
tio.Subject(
    **labels
).plot(figsize=(20,10), output_path='sohee_stent_segmentations')
# # img.get_data_dtype()
```
